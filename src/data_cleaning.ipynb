```python
import pandas as pd
```

```python
data=pd.read_csv("../data/online_retail.csv")
```

```python
data.head(10)
```

### 1. Removing duplicates

```python
data=data.drop_duplicates()
```

```python
data.info()
```

### 2. Removing null values

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

### 3. Converting the Object to Date time 

```python
data['InvoiceDate']=pd.to_datetime(data['InvoiceDate'])
```

```python
data['InvoiceDate']
```

```python
data.info()
```

### 4. Creating a new column Revenue

```python
data['Revenue']=data['Quantity']*data['UnitPrice']
```

```python
data.info()
```

### 5. Removing negative values from quantity

```python
data = data[data['Quantity'] > 0]
```

```python
data.info()
```

### 6. Removing negative values from price

```python
data = data[data['UnitPrice'] > 0]
```

```python
data.info()
```

```python
data['CustomerID'].unique()
```

### 7. Converting CustomerID to Int type

```python
data['CustomerID']=data['CustomerID'].astype(int)
```

```python
data['CustomerID']
```

```python
data.info()
```

```python
data['Country'].unique()
```

### 8. Remove unspecified country entries

```python
data = data[data['Country'] != 'Unspecified']
```

```python
data.info()
```

## Saving the cleaned data

```python
data.to_csv('../data/cleaned_sales_data.csv', index=False)
```
